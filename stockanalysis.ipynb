```python
import numpy as np
import pandas as pd
import yfinance as yf
from skimage import io
from matplotlib import pyplot as plt

tickers_symbols= {
    'Etisalat': ['7020.SR'],
    'AT&T': ['T'],
    'Verizon': ['VZ'],
    'BT':['BT-A.L'],
    'A1 Telekom Austria': ['TKA.VI'],
    'Deutsche Telekom': ['DTE.DE'],
    'Vodafone': ['VOD'],
    'Orange': ['ORA.PA'],
    'Vivendi': ['VIV.PA']
}

count = 1

for i in tickers_symbols:
    tickers_symbols[i].append(yf.Ticker(tickers_symbols[i][0]))
    
for i in tickers_symbols:
    plt.subplot(3,6,count)
    history = tickers_symbols[i][1].history(period='5y')
    plt.plot(history.Close)
    count+=1
    plt.subplot(3,6,count)
    f = tickers_symbols[i][1].info['logo_url']
    a = io.imread(f)
    plt.imshow(a)
    plt.axis('off')
    count+=1
```
